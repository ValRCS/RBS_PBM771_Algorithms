
# 📘 The Akra–Bazzi Theorem — Generalizing Divide and Conquer Recurrences

The **Akra–Bazzi Theorem** is a powerful tool for solving complex **divide-and-conquer recurrences** where the subproblems are *unevenly divided* or *non-uniformly scaled*.

It generalizes the **Master Theorem**, allowing us to analyze recurrences such as:

\[
T(n) = T(2n/3) + T(n/3) + O(1)
\]

or

\[
T(n) = T(2n/3) + T(n/2) + O(1)
\]

which the classical Master Theorem cannot handle directly.



## 🧩 1. The General Form

The Akra–Bazzi theorem handles recurrences of the form:

\[
T(x) = \sum_{i=1}^{k} g_i(x)\, T(h_i(x)) + f(x)
\]

For most divide-and-conquer algorithms, this can be simplified to:

\[
T(x) = \sum_{i=1}^{k} a_i T(b_i x + O(x^\alpha)) + f(x)
\]

where:

- \(a_i > 0\): number (or weight) of subproblems
- \(0 < b_i < 1\): fraction of the problem size for subproblem \(i\)
- \(f(x)\): cost of combining results or other work done outside recursion
- \(\alpha < 1\): small additive shift term allowed inside recursive sizes



## ⚙️ 2. The Akra–Bazzi Theorem Statement

If \(T(x)\) satisfies the above recurrence, then:

\[
T(x) = \Theta\!\left(x^p \left(1 + \int_{1}^{x} \frac{f(u)}{u^{p+1}} \, du \right) \right)
\]

where \(p\) is the **unique real number** satisfying:

\[
\sum_{i=1}^{k} a_i b_i^{p} = 1
\]

This exponent \(p\) plays the same role as \(\log_b a\) in the **Master Theorem**.



## 🧮 3. Step-by-Step Procedure

1. **Identify parameters:** rewrite your recurrence as \(T(n) = \sum_i a_i T(b_i n) + f(n)\)
2. **Solve for \(p\):** find \(p\) that satisfies \(\sum_i a_i b_i^p = 1\)
3. **Plug in \(f(n)\)** to the main formula
4. **Simplify the integral:** for common cases of \(f(n) = O(n^c)\):

| Case | Result |
|------|---------|
| If \(c < p\) | \(T(n) = \Theta(n^p)\) |
| If \(c = p\) | \(T(n) = \Theta(n^p \log n)\) |
| If \(c > p\) | \(T(n) = \Theta(f(n))\) |



## 🧠 4. Why We Need It

The Master Theorem assumes all subproblems are equal in size — for example:

\[
T(n) = aT\left(\frac{n}{b}\right) + f(n)
\]

But Akra–Bazzi handles **unequal or nonuniform splits**, such as:

\[
T(n) = T(2n/3) + T(n/3) + O(1)
\]

where the fractions \(2/3\) and \(1/3\) are not the same.



## 🧩 5. Example 1 — Uneven Split

\[
T(n) = T(2n/3) + T(n/3) + O(1)
\]

**Step 1.** Parameters: \(a_1=a_2=1\), \(b_1=2/3\), \(b_2=1/3\), \(f(n)=O(1)\).

**Step 2.** Solve for \(p\): \((2/3)^p + (1/3)^p = 1\) → \(p = 1\).

**Step 3.** Combine cost \(O(1)\) is smaller than \(n^p\), so:

\[
\boxed{T(n) = \Theta(n)}
\]



## 🧩 6. Example 2 — Biased Split

\[
T(n) = 2T(2n/3) + O(1)
\]

**Step 1.** Parameters: \(a=2\), \(b=2/3\).

**Step 2.** Solve for \(p\): \(2(2/3)^p = 1\) → \(p = \log_{3/2}2 \approx 1.7095\).

**Step 3.** Combine cost is constant, so:

\[
\boxed{T(n) = \Theta(n^{1.7095})}
\]



## 🧩 7. Example 3 — Asymmetric Split

\[
T(n) = T(2n/3) + T(n/2) + O(1)
\]

**Step 1.** Parameters: \(a_1=a_2=1\), \(b_1=2/3\), \(b_2=1/2\).

**Step 2.** Solve for \(p\): \((2/3)^p + (1/2)^p = 1\).

Using numerical approximation, \(p \approx 1.295\).

**Step 3.** Combine cost \(O(1)\) → smaller than \(n^p\).

\[
\boxed{T(n) = \Theta(n^{1.295})}
\]



## 🧩 8. Example 4 — Quickselect Average Case

\[
T(n) = T(n/2) + O(n)
\]

**Step 1.** Parameters: \(a=1\), \(b=1/2\), \(f(n)=O(n)\).

**Step 2.** Solve for \(p\): \((1/2)^p = 1\) → \(p=0\).

**Step 3.** Since \(f(n) = O(n)\) grows faster than \(n^p = 1\):

\[
\boxed{T(n) = \Theta(n)}
\]



## 📊 9. Visualizing \( \sum a_i b_i^p = 1 \)

Let's visualize how we find \(p\) by plotting \(f(p) = \sum_i a_i b_i^p\) and finding where it crosses 1.


```

import numpy as np
import matplotlib.pyplot as plt

def find_p(b, a):
    p_vals = np.linspace(0, 2, 200)
    f_vals = sum(a_i * (b_i ** p_vals) for a_i, b_i in zip(a, b))
    plt.figure(figsize=(6, 4))
    plt.plot(p_vals, f_vals, label='Σ aᵢ bᵢ^p')
    plt.axhline(1, color='red', linestyle='--', label='=1 line')
    plt.xlabel('p')
    plt.ylabel('Σ aᵢ bᵢ^p')
    plt.title('Finding p from Σ aᵢ bᵢ^p = 1')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example for T(n)=T(2n/3)+T(n/2)+O(1)
b = [2/3, 1/2]
a = [1, 1]
find_p(b, a)
```

## ✅ 11. Key Takeaways

- Akra–Bazzi **extends** the Master Theorem to *unequal subproblem sizes*.
- The critical exponent \(p\) is found from \(\sum a_i b_i^p = 1\).
- The method can handle **nonuniform**, **additive-shift**, and **continuous** recurrences.
- When subproblem fractions sum to 1 → \(O(n)\) complexity.
- When they sum to >1 → **superlinear** growth (e.g., \(O(n^{1.7})\)).
- When they sum to <1 → **sublinear** (tree shrinks quickly).

This theorem is especially useful for analyzing recursive geometric algorithms, unbalanced divide-and-conquer methods, and irregular tree-structured computations.
